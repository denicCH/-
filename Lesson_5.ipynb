1. Найти доверительный интервал для оценки математического ожидания $a$ нормального распределения с надежностью 0,95, зная выборочную среднюю $ \overline{x} = 75.12$, объем выборки $n=121$ и среднее квадратическое отклонение $\sigma=11$.

Найдем доверительный интервал для математического ожидания a с
надежностью 0,95, используя формулу: 
$$\overline{x} - c_{\gamma}\cdot \frac{\sigma}{\sqrt{n}} < a < \overline{x} + c_{\gamma}\cdot \frac{\sigma}{\sqrt{n}}$$

где $c_{\gamma}$ определяется из таблицы (обратное значение функции стандартного нормального распределения) из условия:
$$Ф(c_{\gamma}) = \frac{\gamma}{2} = 0.95/2 = 0.475$$
$$c_{\gamma} \approx 1.96$$

Получаем после подстановки известных данных: 

$$75.12 - 1.96 \cdot \frac{{11}}{{\sqrt {121} }} < a < 75.12 + 1.96 \cdot \frac{{11}}{{\sqrt {121} }}$$

$$73.16 < a < 77.08$$

2. С целью размещения рекламы опрошено 600 телезрителей, из которых данную передачу смотрят 170 человек. С доверительной вероятностью 0,91 найти долю телезрителей, охваченных рекламой в наихудшем случае.

Для оценки неизвестной доли телезрителей используем формулу: 
$$w - t\sqrt{\frac{w(1-w)}{n}} < p < w + t\sqrt{\frac{w(1-w)}{n}}$$
где n - объём выборки, $w = \frac{170}{600} = \frac{17}{60}$ - относительная частота события, $t = Ф^{-1} (\gamma/2) = Ф^{-1} (0.91/2) = 1.69$  (из таблицы функции Лапласа)

Подставляем данные:

$$\frac{{17}}{{60}} - 1.69\sqrt {\frac{{\frac{{17}}{{60}}\left( {1 - \frac{{17}}{{60}}} \right)}}{{600}}}  < p < \frac{{17}}{{60}} + 1.69\sqrt {\frac{{\frac{{17}}{{60}}\left( {1 - \frac{{17}}{{60}}} \right)}}{{600}}} $$

$$0.252 < a < 0.314$$

Таким образом, получаем что худший вариант охвата телезрителей – 25.2% телезрителей.
